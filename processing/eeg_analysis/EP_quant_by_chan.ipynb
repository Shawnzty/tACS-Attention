```python
case = 'endo valid'
watch = '11 stim'
tmin = -0.1
tmax = 0.5
real_ids = [1, 3, 4, 5, 9, 12, 13, 17, 18]
sham_ids = [2, 6, 7, 8, 10, 11, 14, 15, 16]

highpass = 0.3
lowpass = 30

sham_evoked_before, sham_evoked_after, real_evoked_before, real_evoked_after, _, _ = fe.pipeline_EP_allsubs(case, watch, tmin, tmax, hipass=highpass, lopass=lowpass)
```

```python
# evaluate all channels
pass_rate = []
windows = {
                "N75": [0.055, 0.085],
                "P100": [0.080, 0.110],
                "N145": [0.130, 0.160]
            }

for session_data in [sham_evoked_before, sham_evoked_after, real_evoked_before, real_evoked_after]:
    for group_id in range (len(real_evoked_before)):
        data = session_data[group_id]
        time_vector = np.linspace(tmin, tmax, data.shape[2])
        # onesub_EP = np.empty((data.shape[0], data.shape[1], 2, 3))
        onesub_EP = []

        for i in range(data.shape[0]):
            for j in range(data.shape[1]):
                one_trial = data[i,j,:]
                peaks = fe.detect_EP(one_trial, time_vector, windows)
                if not np.isnan(peaks).any():
                    onesub_EP.append(peaks)
        
        pass_rate.append(len(onesub_EP)/(data.shape[0]*data.shape[1]))

print(sum(pass_rate)/len(pass_rate))
```

```python
# plot one trial and see
# Define sample data
data_points = data[40,20,:]  # random data points

# Define time vector
sampling_freq = 1200
time_vector = np.linspace(tmin, tmax, int((tmax-tmin)*sampling_freq + 1))

colors = ["red", "green", "blue"]

# Plot the data points
plt.plot(time_vector, data_points, label="EEG Signal")

# Highlight time windows
for idx, (key, window) in enumerate(windows.items()):
    plt.axvspan(window[0], window[1], color=colors[idx], alpha=0.3, label=key)

# Add legend and labels
plt.legend(loc="upper right")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("EEG Signal with Time Windows")

plt.tight_layout()
plt.show()
```

```python
# container
sham_before = []
sham_after = []
real_before = []
real_after = []
data_list = [sham_evoked_before, sham_evoked_after, real_evoked_before, real_evoked_after]
know_list = [[], [], [], []]

# choose channels
bad_channels = [
        [ # sham before
            [], [], [], [], [], [], [22,21], [5,9], []
        ],
        [ # sham after
            [], [], [], [], [], [], [], [], []
        ],
        [ # real before
            [], [], [], [], [], [], [], [], []
        ],
        [ # real after
            [], [], [], [], [], [7], [], [], []
        ]
]


# for groups and sessions
for i, session_data in enumerate(data_list):
    # for channels
    for j in range(32):  # channels
        one_chan = []
        # for subjects in group/session
        for group_id in range(9):
            if j+1 not in bad_channels[i][group_id]:
                evoked_data = session_data[group_id]
                time_vector = np.linspace(tmin, tmax, evoked_data.shape[2])
                # for trials
                for k in range(evoked_data.shape[0]):  # trials
                    one_trial = evoked_data[k,j,:]
                    peaks = fe.detect_EP(one_trial, time_vector, windows)
                    if not np.isnan(peaks).any():
                        one_chan.append(peaks)
                
        know_list[i].append(one_chan)
```

```python
def plot_boxplot(sham_before, sham_after, real_before, real_after, channel):
    fig, axes = plt.subplots(2, 3, figsize=(10, 7))
    labels = ['Sham Before', 'Sham After', 'Real Before', 'Real After']
    colors = ['lightblue', 'blue', 'lightcoral', 'red']
    ylim_factor = [[0.065, 0.075],[0.09, 0.1],[0.14, 0.15]]
    channel_dict = {'Fp1':1, 'Fp2':2, 'AF3':3, 'AF4':4, 'F7':5, 'F3':6, 'Fz':7, 'F4':8, 'F8':9, 'FC1':10, 
                'FC2':11, 'T7':12, 'C3':13, 'Cz':14, 'C4':15, 'T8':16, 'CP5':17, 'CP1':18, 'CP2':19, 'CP6':20, 
                'P7':21, 'P5':22, 'P3':23, 'Pz':24, 'P4':25, 'P6':26, 'P8':27, 'PO3':28, 'PO4':29, 'O1':30, 'Oz':31, 'O2':32}
    titles = [['N75 Latency','P100 Latency', 'N145 Latency'],['N75 Amplitude','P100 Amplitude', 'N145 Amplitude']]

    for i in range(2):
        for j in range(3):
            # Extract data for the subplot
            sham1 = [sublist[i][j] for sublist in sham_before]
            sham2 = [sublist[i][j] for sublist in sham_after]
            real1 = [sublist[i][j] for sublist in real_before]
            real2 = [sublist[i][j] for sublist in real_after]

            sham1 = fe.rm_outlier(sham1, left_k=1.5, right_k=1.5, verbose=True)
            sham2 = fe.rm_outlier(sham2, left_k=1.5, right_k=1.5)
            real1 = fe.rm_outlier(real1, left_k=1.5, right_k=1.5)
            real2 = fe.rm_outlier(real2, left_k=1.5, right_k=1.5)

            # Compute p-values
            p1 = mannwhitneyu(sham1, sham2).pvalue * 4
            p2 = mannwhitneyu(real1, real2).pvalue * 4
            p3 = mannwhitneyu(sham1, real1).pvalue * 4
            p4 = mannwhitneyu(sham2, real2).pvalue * 4

            # Bar heights (means) and error (standard deviations)
            means = [np.mean(data) for data in [sham1, sham2, real1, real2]]
            std_errors = [
                np.std(sham1) / np.sqrt(len(sham1)), np.std(sham2) / np.sqrt(len(sham2)),
                np.std(real1) / np.sqrt(len(real1)), np.std(real2) / np.sqrt(len(real2))
            ]

            # Bar plot
            x_positions = [1, 2, 3, 4]
            axes[i, j].bar(x_positions, means, yerr=std_errors, color=colors, capsize=5, align='center', width=0.5)

            # Setting x-axis labels and title
            axes[i, j].set_xticks(x_positions)
            axes[i, j].set_xticklabels(labels)
            axes[i, j].set_title(f'{titles[i][j]}')
            
            # Display p-values on the plot
            if i == 1 and (j==0 or j==2):
                axes[i, j].text(1.5, 1.05 * min(means), f'p={p1:.3f}', ha='center')
                axes[i, j].text(3.5, 1.05 * min(means), f'p={p2:.3f}', ha='center')
                axes[i, j].text(2, 1.15 * min(means), f'p={p3:.3f}', ha='center')
                axes[i, j].text(3, 1.15 * min(means), f'p={p4:.3f}', ha='center')
            elif i == 1 and j == 1:
                axes[i, j].text(1.5, 1.05 * max(means), f'p={p1:.3f}', ha='center')
                axes[i, j].text(3.5, 1.05 * max(means), f'p={p2:.3f}', ha='center')
                axes[i, j].text(2, 1.15 * max(means), f'p={p3:.3f}', ha='center')
                axes[i, j].text(3, 1.15 * max(means), f'p={p4:.3f}', ha='center')
            else:
                axes[i, j].text(1.5, 1.01 * max(means), f'p={p1:.3f}', ha='center')
                axes[i, j].text(3.5, 1.01 * max(means), f'p={p2:.3f}', ha='center')
                axes[i, j].text(2, 1.02 * max(means), f'p={p3:.3f}', ha='center')
                axes[i, j].text(3, 1.02 * max(means), f'p={p4:.3f}', ha='center')

            # Setting y-axis limits
            axes[0, j].set_ylim(ylim_factor[j])
    
    # Main title for each figure
    channel_name = [key for key, value in channel_dict.items() if value == channel][0]
    fig.suptitle(f"Channel: {channel_name}", fontsize=16, y=1.05)
    
    plt.tight_layout()
    plt.show()



# Call the function
for channel in range(1,33):
    sham_before = know_list[0][channel-1]
    sham_after = know_list[1][channel-1]
    real_before = know_list[2][channel-1]
    real_after = know_list[3][channel-1]
    
    plot_boxplot(sham_before, sham_after, real_before, real_after, channel)
```

```python
# Sham
image_paths = []
for group_id in range(9):
    evokeds = {"Before": sham_evoked_before[group_id], "After": sham_evoked_after[group_id]}
    
    picks = list(channels.keys())
    
    picks = [item for item in picks if item not in bad_channels[0][group_id]]
    
    # Save the plot to a file
    fig = mne.viz.plot_compare_evokeds(evokeds, combine="mean", title=case + ", sham: " + str(sham_ids[group_id]), 
                                       show_sensors="lower right", picks=picks, show=False, colors=['lightblue', 'blue'])
    
    image_path = os.path.join('tmp_files', f"temp_plot_{group_id}.png")
    fig[0].savefig(image_path)
    image_paths.append(image_path)  # This line is corrected
    plt.close(fig[0])

# Initialize the 3x3 grid of subplots
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

for idx, ax in enumerate(axs.ravel()):
    img = mpimg.imread(image_paths[idx])
    ax.imshow(img)
    ax.axis('off')  # Turn off axis numbers and ticks


# Adjust the spacing between the subplots
plt.subplots_adjust(wspace=0, hspace=0)

plt.show()

for path in image_paths:
    os.remove(path)
```
